**use case-1**

- create a empty data frame first

- create 3 list usin list comprehension

- list1=a values from 1 to 10

- list2=square of 1 to 10

- list3=cube of 1 to 10

- finally make a data frame

```python
import pandas as pd
list1=[i for i in range(1,11) ]
list2=[i**2 for i in range(1,11)]
list3=[i**3 for i in range(1,11)]
cols=["values","square","cube"]
pd.DataFrame(zip(list1,list2,list3),columns=cols)
```

- if you want to drop a specific column 

- we will use dropo method

- we already know about difference between keywords and methods

- keywords will use directly 

- methods will use with corresponding dataframe

- **method name: drop**

- column = which column you want to drop 

- axis : represents the specified value is a row or column 

    - axis = 0 means rows 

    - axis = 1 means columns
 
    - by default axis = 0
 
- inplace : will provide overwrite the dataframe or not

    - inplace = True , this indicates overwrite the dataframe
 
    - inplace = Flase , this indicates the modified dataframe will be use when you provide new variable for the operation
      

```python
df_new=df1.drop("z",axis=1)
df_new
# if you want to use then we need to assign a variable
```

```python
df1.drop("y",axis=1,inplace=True)
df1
```
